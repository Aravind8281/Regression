```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
np.random.seed(42)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
y = 4 + 3 * X1 + 2 * X2 + np.random.randn(100, 1)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1, X2, y, c='blue', marker='o')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
plt.title('Generated Dataset for Multiple Linear Regression')
plt.show()
X_multi = np.hstack((X1, X2))
model_multi = LinearRegression()
model_multi.fit(X_multi, y)
X1_new = np.array([[1.5]])
X2_new = np.array([[2.5]])
X_new_multi = np.hstack((X1_new, X2_new))
y_pred_multi = model_multi.predict(X_new_multi)
print("Intercept:", model_multi.intercept_[0])
print("Coefficients (slope):", model_multi.coef_[0])
y_pred_train_multi = model_multi.predict(X_multi)
mse_multi = mean_squared_error(y, y_pred_train_multi)
print("Mean Squared Error:", mse_multi)
```
