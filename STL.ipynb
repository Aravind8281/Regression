```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL

# Generate some example time series data with seasonality and trend
np.random.seed(42)
time = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
data = np.sin(2 * np.pi * (time.dayofyear) / 365) + np.random.normal(scale=0.2, size=len(time))
ts = pd.Series(data, index=time)

# Perform STL decomposition
stl = STL(ts, seasonal=13)  # 13 for yearly seasonality, adjust as needed
result = stl.fit()

# Plot the original time series and the decomposed components
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(ts, label='Original Time Series')
plt.title('Original Time Series')

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.title('Trend Component')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.title('Seasonal Component')

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residual')
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```
